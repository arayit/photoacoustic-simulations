# exp_001 — EDA: TPA-PA with BODIPY-TR

Exploratory analysis of photoacoustic signal generation via two-photon absorption.

**Scenarios:**
- `s01`: NS single pulse (3 ns, 1 J/cm²)
- `s02`: FS single pulse (100 fs, 0.1 J/cm²)
- `s03_burst_N010...N300`: FS burst mode (N = 10–300 pulses)

```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

plt.style.use('dark_background')
plt.rcParams['figure.figsize'] = [12, 6]
plt.rcParams['figure.dpi'] = 100
```

## 1. Load all results

```python
results_dir = Path(r'C:\Users\arayi\Documents\MATLAB\exp_001\results')

def load_mat(fpath):
    """Load key fields from v7.3 .mat file."""
    data = {}
    with h5py.File(fpath, 'r') as f:
        r = f['results']
        # Sensor data: stored as (Nt, 128), transpose to (128, Nt)
        data['sensor_data'] = r['sensor_data'][:].T
        data['t_array'] = r['t_array'][:].flatten()
        # p0: stored as (Nz, Ny) in HDF5, need to transpose for (Ny, Nz)
        data['p0_acoustic'] = r['p0_acoustic'][:].T
        data['y_vec'] = r['y_vec'][:].flatten()
        data['z_vec'] = r['z_vec'][:].flatten()
        # Config
        cfg = r['cfg']
        data['pulse_duration'] = cfg['pulse_duration'][0, 0]
        data['fluence_focus'] = cfg['fluence_focus'][0, 0]
        data['alpha2_target'] = cfg['alpha2_target'][0, 0]
        data['target_depth'] = cfg['target_depth'][0, 0]
    return data

# Load all files
files = sorted(results_dir.glob('*.mat'))
print(f"Found {len(files)} result files")

results = {}
for f in files:
    label = f.stem
    results[label] = load_mat(f)

# Check shapes
d = results['s01_ns_tau3ns_F1']
print(f"p0 shape: {d['p0_acoustic'].shape}")
print(f"y_vec: {d['y_vec'].shape}, z_vec: {d['z_vec'].shape}")
```

## 2. Extract summary metrics

```python
# Build summary table
summary = []

for label, d in results.items():
    sensor = d['sensor_data']
    p_max = np.max(np.abs(sensor))
    central_trace = sensor[63, :]
    p_center_max = np.max(np.abs(central_trace))
    
    if 'burst' in label:
        N = int(label.split('_N')[1])
    else:
        N = 1
    
    summary.append({
        'label': label,
        'N': N,
        'tau_s': d['pulse_duration'],
        'F_Jcm2': d['fluence_focus'],
        'p_max_Pa': p_max,
        'p_center_max_Pa': p_center_max,
    })

import pandas as pd
df = pd.DataFrame(summary)
df = df.sort_values('N').reset_index(drop=True)
df
```

## 3. Compare NS vs FS vs Burst

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Left: Peak pressure vs N
ax = axes[0]
burst_df = df[df['label'].str.contains('burst')]
ax.plot(burst_df['N'], burst_df['p_max_Pa'] * 1e-3, 'o-', color='cyan', label='Burst (max)')
ax.plot(burst_df['N'], burst_df['p_center_max_Pa'] * 1e-3, 's--', color='orange', label='Burst (center)')

ns_p = df[df['label'] == 's01_ns_tau3ns_F1']['p_max_Pa'].values[0] * 1e-3
fs_p = df[df['label'] == 's02_fs_tau100fs_F01']['p_max_Pa'].values[0] * 1e-3
ax.axhline(ns_p, color='red', ls=':', lw=2, label=f'NS: {ns_p:.4f} kPa')
ax.axhline(fs_p, color='lime', ls=':', lw=2, label=f'FS: {fs_p:.4f} kPa')

ax.set_xlabel('Burst N')
ax.set_ylabel('Peak pressure [kPa]')
ax.set_title('PA Signal vs Burst Size')
ax.legend(loc='upper left')
ax.grid(alpha=0.3)

# Right: Peak pressure vs fluence
ax = axes[1]
ax.plot(burst_df['F_Jcm2'], burst_df['p_max_Pa'] * 1e-3, 'o-', color='cyan', label='Burst')
ax.scatter([1.0], [ns_p], color='red', s=100, marker='*', zorder=5, label='NS')
ax.scatter([0.1], [fs_p], color='lime', s=100, marker='*', zorder=5, label='FS')

ax.set_xlabel('Total fluence [J/cm²]')
ax.set_ylabel('Peak pressure [kPa]')
ax.set_title('PA Signal vs Total Fluence')
ax.legend()
ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

## 4. Time traces comparison

```python
fig, axes = plt.subplots(3, 1, figsize=(12, 8), sharex=True)

scenarios = ['s01_ns_tau3ns_F1', 's02_fs_tau100fs_F01', 's03_burst_N100']
colors = ['red', 'lime', 'cyan']
titles = ['NS (3 ns, 1 J/cm²)', 'FS (100 fs, 0.1 J/cm²)', 'Burst (N=100, 10 J/cm²)']

for ax, label, color, title in zip(axes, scenarios, colors, titles):
    d = results[label]
    t_us = d['t_array'] * 1e6
    trace = d['sensor_data'][63, :]
    
    ax.plot(t_us, trace * 1e-3, color=color, lw=0.8)
    ax.set_ylabel('Pressure [kPa]')
    ax.set_title(title)
    ax.grid(alpha=0.3)

axes[-1].set_xlabel('Time [µs]')
plt.tight_layout()
plt.show()
```

## 5. Initial pressure (p0) maps

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

scenarios = ['s01_ns_tau3ns_F1', 's02_fs_tau100fs_F01', 's03_burst_N100']
titles = ['NS (3 ns)', 'FS (100 fs)', 'Burst (N=100)']

for ax, label, title in zip(axes, scenarios, titles):
    d = results[label]
    p0 = d['p0_acoustic']
    y_mm = d['y_vec'] * 1e3
    z_mm = d['z_vec'] * 1e3
    
    # Find ROI around focus
    z_focus = d['target_depth'] * 1e3
    z_idx = np.where((z_mm > z_focus - 0.5) & (z_mm < z_focus + 0.5))[0]
    y_idx = np.where((y_mm > -0.1) & (y_mm < 0.1))[0]
    
    p0_roi = p0[np.ix_(y_idx, z_idx)]
    y_roi = y_mm[y_idx]
    z_roi = z_mm[z_idx]
    
    im = ax.imshow(p0_roi * 1e-3, 
                   extent=[z_roi[0], z_roi[-1], y_roi[-1], y_roi[0]],
                   aspect='auto', cmap='hot')
    ax.set_xlabel('z [mm]')
    ax.set_ylabel('y [mm]')
    ax.set_title(f'{title}\nmax = {np.max(p0_roi)*1e-3:.4f} kPa')
    plt.colorbar(im, ax=ax, label='p0 [kPa]')

plt.tight_layout()
plt.show()
```

## 6. TPA scaling: p_max vs N

```python
fig, ax = plt.subplots(figsize=(8, 6))

N_vals = burst_df['N'].values
p_vals = burst_df['p_max_Pa'].values

coeffs = np.polyfit(N_vals, p_vals, 1)
p_fit = np.polyval(coeffs, N_vals)

ax.plot(N_vals, p_vals * 1e-3, 'o', color='cyan', ms=8, label='Data')
ax.plot(N_vals, p_fit * 1e-3, '--', color='yellow', lw=2, 
        label=f'Linear fit (slope={coeffs[0]*1e-6:.2f} kPa/pulse)')

ax.set_xlabel('N (number of pulses)')
ax.set_ylabel('Peak pressure [kPa]')
ax.set_title('TPA Scaling: Burst Mode')
ax.legend()
ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()

print(f"Linear fit: p = {coeffs[0]*1e-6:.4f} kPa/pulse * N + {coeffs[1]*1e-3:.4f} kPa")
```

## 7. B-mode image (stacked sensor data)

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

scenarios = ['s01_ns_tau3ns_F1', 's02_fs_tau100fs_F01', 's03_burst_N100']
titles = ['NS', 'FS', 'Burst N=100']

for ax, label, title in zip(axes, scenarios, titles):
    d = results[label]
    sensor = d['sensor_data']
    t_us = d['t_array'] * 1e6
    
    im = ax.imshow(np.abs(sensor), 
                   extent=[t_us[0], t_us[-1], 127, 0],
                   aspect='auto', cmap='hot')
    ax.set_xlabel('Time [µs]')
    ax.set_ylabel('Element #')
    ax.set_title(title)
    plt.colorbar(im, ax=ax, label='|p| [Pa]')

plt.tight_layout()
plt.show()
```

## 8. Summary statistics

```python
ns = results['s01_ns_tau3ns_F1']
fs = results['s02_fs_tau100fs_F01']
burst100 = results['s03_burst_N100']

print("=== Peak Pressure Comparison ===")
print(f"NS (3ns, 1 J/cm²):        {np.max(np.abs(ns['sensor_data']))*1e-3:.4f} kPa")
print(f"FS (100fs, 0.1 J/cm²):    {np.max(np.abs(fs['sensor_data']))*1e-3:.4f} kPa")
print(f"Burst N=100 (10 J/cm²):   {np.max(np.abs(burst100['sensor_data']))*1e-3:.4f} kPa")
print()

I_ns = 1 / 3e-9
I_fs = 0.1 / 100e-15
print("=== Peak Intensity ===")
print(f"NS:  {I_ns:.2e} W/cm²")
print(f"FS:  {I_fs:.2e} W/cm²")
print(f"Ratio FS/NS: {I_fs/I_ns:.0f}x")
```
